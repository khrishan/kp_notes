## Python Setup for Windows

### Install Python

1. Go to [http://python.org](python.org)
1. Hover over `Downloads` and click on the `latest version`.
1. Click `Install Now` (and allow it to install)
1. Verify that it is installed. Open a command prompt, and type `python`.
    + If the above doesn't work, then you may need to reboot your machine.

![Windows Command Prompt with Python 3.6 showing](img/00_python_windows_cmd.png)

There is an example chart that can be run [exchange.py](examples/exchange.py). This module will show you the exchange rates for three currencies, over the last 12 months, and draw a graph.

### Install IntelliJ

1. Visit [jetbrains.com](https://www.jetbrains.com)
1. Download the `community version` for the IDE. Then run through the installaion wizard.
1. First time you open IntelliJ, it will ask if you want to import settings. This is only useful if you are a previous user of IntelliJ.

#### How to Install Plugin
1. Click `configure`, `plugins`
1. A new pop up window will appear. Search for `python`. Install the `Jet Brains` plugin. (`Python Community Edition`).
1. Make sure you check the `Add Python3.6 to path`
1. Click install.
1. Restart IntelliJ

#### Configure Intellij

##### Enable Python
*This process may not be required, as it may have sorted itself our already.*
1. `Configure, Project Defaults, Project Structure, SDKs, (click) +, Python SDK`
    + Under the `Virtualenv` section, make sure that `Python3` is enabled.

##### Enable Line Numbers
1. `Configure, Settings, General, Apperance`. Tick box to `Enable Line Numbers`.

## Python Setup for Mac

### Install Python

1. Go to [http://python.org](python.org).
1. Hover over `Downloads` and click on the `latest version`.
1. Run the download file like a normal installation wizard.
1. Check all has been installed by going to terminal and typing `python3`.

There is an example chart that can be run [exchange.py](examples/exchange.py). This module will show you the exchange rates for three currencies, over the last 12 months, and draw a graph.

### Install IntelliJ

1. Visit [jetbrains.com](https://www.jetbrains.com)
1. Download the `community version` for the IDE. Then run through the installaion wizard.
1. First time you open IntelliJ, it will ask if you want to import settings. This is only useful if you are a previous user of IntelliJ.

#### How to Install Plugin
1. Click `configure`, `plugins`
1. A new pop up window will appear. Search for `python`. Install the `Jet Brains` plugin. (`Python Community Edition`).
1. Make sure you check the `Add Python3.6 to path`
1. Click install.
1. Restart IntelliJ

#### Configure Intellij

##### Enable Python
*This process may not be required, as it may have sorted itself our already.*
1. `Configure, Project Defaults, Project Structure, SDKs, (click) +, Python SDK`
    + Under the `Virtualenv` section, make sure that `Python3` is enabled.
1. Select the two checkboxes (`Inherit Global Site-Packages` and `Make Available to all Projects`)

##### Enable Line Numbers
1. `Configure, Settings, General, Apperance`. Tick box to `Enable Line Numbers`.

## Python Setup for Linux

Normally Python2 and Python3 are already installed. To check, type `python` and `python3` into the termimal.

### Install IntelliJ

1. Visit [jetbrains.com](https://www.jetbrains.com)
1. Download the `community version` for the IDE. Then run through the installaion wizard.
1. Go to the `Downloads` folder, and extract the files out of the `tar` file to the location that you want it to be in.
1. Open `terminal` and navigate to the extracted files and then the `bin` folder.
1. There should be a file called `idea.sh`. To run, use the command `./idea.sh`.
1. You can now find IntelliJ through the Applications Menu.
1. First time you open IntelliJ, it will ask if you want to import settings. This is only useful if you are a previous user of IntelliJ.

#### How to Install Plugin
1. Click `configure`, `plugins`
1. A new pop up window will appear. Search for `python`. Install the `Jet Brains` plugin. (`Python Community Edition`).
1. Make sure you check the `Add Python3.6 to path`
1. Click `Install`.
1. Restart IntelliJ

#### Configure Intellij

##### Enable Python
*This process may not be required, as it may have sorted itself our already.*
1. `Configure, Project Defaults, Project Structure, SDKs, (click) +, Python SDK`
    + Under the `Virtualenv` section, make sure that `Python3` is enabled.
1. Select the two checkboxes (`Inherit Global Site-Packages` and `Make Available to all Projects`)

##### Enable Line Numbers
1. `Configure, Settings, General, Apperance`. Tick box to `Enable Line Numbers`.
