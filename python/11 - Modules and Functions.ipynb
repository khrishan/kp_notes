# Times & Datetime

There are two 'pitfalls' when dealing with dates and time. **Time zones** and **Daylight Saving Times**.

Best practice is to save time in 'UTC - Universal Time Coordinated' (comprimise with the French)' - with an offset due to a paricular timezone.

Python has three modules - as part of the standard library to help us.
+ [`time`](https://docs.python.org/3/library/time.html)     - good when dealing with just elapsed time.
+ `datetime` -
+ `calendar` -

`epoch` - January 1st 1970 -  x86 OS will stop working properly in Feburary 2038 (should be off the grid by then)

```python
import time

print(time.gmtime(0)) # Always works in UTC = converts number of seconds to tuple.
print(time.localtime())
print(time.time())
```

```python
time_here = time.localtime()
print(time_here)

print('Year : {} {}'.format(time_here[0], time_here.tm_year))
print('Month : {} {}'.format(time_here[1], time_here.tm_mon))
print('Day : {} {}'.format(time_here[2], time_here.tm_mday))
```

```python
# Time Game
import random
import time
#from time import time as my_timer
from time import perf_counter as my_timer #more accurate but cannot convert to local time.
#from time import monotonic as my_timer # similar to perf_counter, but less resolution (time can't go backwards) #don't worry about it
#from time import process_time as my_timer # returns the time the CPU spends on this process # not good for this game

input('Press enter to start')

wait_time = random.randint(1,6)
time.sleep(wait_time)
start_time = my_timer()

input('Press enter to stop')

end_time = my_timer()

print('Started Time  : {}'.format(time.strftime('%X', time.localtime(start_time))))
print('End Time      : {}'.format(time.strftime('%X', time.localtime(end_time))))
print()
print('Reaction Time : {}'.format(end_time - start_time))
```

Write a small program to display information on the four clocks whose functions we have just looked at: i.e `time()`, `perf_counter()`, `monotonic()` and `process_time()`.

```python
import time

print('time()\t\t\t{}'.format(time.get_clock_info('time')))
print('perf_counter()\t{}'.format(time.get_clock_info('perf_counter')))
print('monotonic()\t{}'.format(time.get_clock_info('monotonic')))
print('process_time()\t{}'.format(time.get_clock_info('process_time')))
```
