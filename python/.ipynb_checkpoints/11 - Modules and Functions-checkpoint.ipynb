Python has three modules - as part of the standard library to help us.
+ [`time`](https://docs.python.org/3/library/time.html)     - good when dealing with just elapsed time.
+ `datetime` -
+ `calendar` -

````python
import time

print(time.gmtime(0)) # Always works in UTC = converts number of seconds to tuple.
print(time.localtime())
print(time.time())
````

## Challenge - Timezones

Create a program that allows a user to choose one of up to 9 time zones from a menu. You can choose any zones you want from the `all_timezones` list.

The program will then display the time in that timezone, as well as local time and UTC time.

Entering `0` as the choice will quit the program.

Display the dates and times in a format suitable for the user of your prgram to understnad, and include the timezone name when displaying the chosen time.

````python
import datetime
import pytz

available_zones = {'1': 'Africa/Tunis',
                   '2': 'Asia/Kolkata',
                   '3': 'Australia/Adelaide',
                   '4': 'Europe/Brussels',
                   '5': 'Europe/London',
                   '6': 'Japan',
                   '7': 'Pacific/Tahiti',
                   '8': 'US/Hawaii',
                   '9': 'Zulu',
                  }

print('Please choose a timezone (or 0 to quit) : ')
for place in sorted(available_zones):
    print('\t{}. {}'.format(place, available_zones[place]))
    
while True:
    choice = input()
    
    if choice == '0':
        break
    elif choice in available_zones.keys():
        tz_to_display = pytz.timezone(available_zones[choice])
        world_time = datetime.datetime.now(tz=tz_to_display)
        print('The time in {} is {} {}.'.format(available_zones[choice], world_time.strftime('%A %x %X %z'), world_time.tzname()))
        print('Local time is {}.'.format(datetime.datetime.now().strftime('%A %x %X')))
        print('UTC time is {}.'.format(datetime.datetime.utcnow().strftime('%A %x %X')))
        print('')
    else:
        print('That key does not exist. Please try again.')
        print('')
````

````python
# You can also have named parameters, that you don't have to provide values when you call the method

def center_text(*args, sep=' ', end='\n', file=None, flush=False):
    text = ''
    for a in args:
        text += str(a) + ' '

    left_margin = (80 - len(text)) // 2
    print(' ' * left_margin, text, sep=sep, end=end, file=file, flush=flush)

center_text(12)
center_text('Spam and Eggs')
center_text('Spam, Spam and Eggs')
center_text('Spam, Spam, Spam and Spam')
center_text('first', 'second', 2, 3, 4)

# This is how you write the text to a file

with open('centered.txt', mode='w') as c_file :
    center_text(12, file=c_file)
    center_text('Spam and Eggs', file=c_file)
    center_text('Spam, Spam and Eggs', file=c_file)
    center_text('Spam, Spam, Spam and Spam', file=c_file)
    center_text('first', 'second', 2, 3, 4, file=c_file)
````

````python
# this is to test that it has actually been written to a file.

with open('centered.txt', mode='r') as c_file:
    for line in c_file.readlines():
        print(line, end='')
````

## Parabola - More on Functions

````python
# The video used tkinter, but I chose to use matplotlib instead.
# Seeing as I am using jupyter

from matplotlib import pyplot as plt
import numpy as np

%matplotlib inline

fig = plt.figure()

def parabola(x):
    y = x*x
    return y

def print_graph():
    x = []
    y = []
    for i in range(-1000, 1000):
        x.append(i)
        y.append(parabola(i))

    ax = fig.add_axes([0,0,1,1])
    ax.plot(x,y)
    ax.set_title('This is a paraoblic graph!')
    ax.set_xlabel('x')
    ax.set_xlabel('y')
    
print_graph()
````

````python
#Python will change global varable to local variable when not used in the global scope, creating what is called a 'shadow variable'

# In order to change the value of a global variable add the `global` scope at the beginning on the method

x = 0
y = 1

def change_values():
    global x
    global y
    
    x += 5
    y += 6

    
print(x)
print(y)
change_values()
print(x)
print(y)
````

## Importing and Underscores

Convention is to use an underscore `_` when you have a variable that has the same name as a keyword (e.g. `from_`)

In addition, if you put underscores before a method name (whilst it is defined), it becomes a 'protected' method (you can still run it, but your IDE will give you warning telling you the issue.

`from blackjack import *`
Using that `import` statement, Python knows not to import any method that is protected (by the `_`)

`_` can also be a valid variable name, and is known as a 'throwaway variable`.

## Scope and Namespaces

Python allows you to call a function within another function. This is called `recursion`.

A `Recursive` function is one that calls itself.

````python
## Without Recursion

def fact(n):
    """ calculate n! interactively """
    result = 1
    if n > 1:
        for f in range(2, n+1):
            result *= f
    return result


for i in range(130):
    print(i, fact(i))
````

````python
## With Recursion

def fact(n):
    # n! can also be defined as n * (n-1)!
    """ calculate n! recursively """
    
    if n <= 1: # Always have a stopping clause at the start.
        return 1
    else:
        return n * fact(n-1)

for i in range(130):
    print(i, fact(i))
````

````python
# Fibonacci Sequence
# This method takes a lot longer than if we didn't use recursion.

def fib(n):
    """ f(n) = f(n-1) + f(n-2) """
    if n <= 1:
        return n
    else:
        return fib(n-1) + fib(n-2)
    
for i in range(36):
    print(i, fib(i))
````

## OS Filesystem Recursion

````python
import os

listing = os.walk('.')
for root, directories, files in listing:
    print(root)
    for d in directories:
        print(d)
    for f in files:
        print(f)
````

Let's create our own to we can understand what `os.walk()` is doing + understand `recursion` better.

````python
import os


def list_directories(s):
    
    def dir_list(d):
        nonlocal tab_stop # This is new in Python3, which makes it work.
        files = os.listdir(d)
        for f in files:
            current_dir = os.path.join(d, f)
            if os.path.isdir(current_dir):
                print('\t' * tab_stop + 'directory ' + f)
                tab_stop +=1
                dir_list(current_dir)
                tab_stop -=1
            else:
                print('\t' * tab_stop + f)

                
    tab_stop = 0
    if os.path.exists(s):
        print("Directory listing of {}".format(s))
        dir_list(s)
    else:
        print('{} does not exist'.format(s))
        
list_directories('.')
````

## Nonlocals Free and LEGB

More information can be found at : 

[https://docs.python.org/3/library/functions.html#locals](https://docs.python.org/3/library/functions.html#locals)

`global` tells Pythont to look for a variable in the `global` scope

`nonlocal` tells Python to look for said variable in the enclosing fucntion. If it doesn't find it there, it will check in an function closing that and so on. It will **NOT** `global` namespace

The things that create scopes in Python are `modules`, `functions` and `classes`.

At the `module` level, the `local` scope is the same as the `global` scope.

It's helpful to remember the `LEGB` acronym, the order in which Python will search for variables. It stands for 
+ `L`ocal
+ `E`nclosing
+ `G`lobal
+ `B`uilt ins

````python
# Very rare to see this kind of nesting, but this is just an example.

def spam1():
    
    def spam2():
        
        def spam3():
            # Can refer to x & y
            z = ' even' + y
            print('Spam 3 Locals : {}'.format(locals()))
            return z
        
        # Can refer to x
        y = ' more ' + x # y must exist before spam3() is called
        y += spam3()
        print('Spam 2 Locals : {}'.format(locals()))
        return y
    
    x = 'spam' # x must exist before spam2() is called
    x = x + spam2()
    print('Spam 1 Locals : {}'.format(locals()))
    return x

print(spam1())

#print(locals()) 
#print(globals()) # They're identical, trust me!
````
